```python
# Step 0: Install required libraries (uncomment if needed)
!pip install transformers torch sentencepiece datasets evaluate sacrebleu --quiet

# Step 1: Import necessary libraries
from transformers import MarianMTModel, MarianTokenizer
import evaluate # Changed from 'from datasets import load_metric'

# Step 2: Load pretrained MarianMT model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-fr"  # English to French
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Step 3: Prepare source text for translation
src_texts = [
    "Hello, how are you?",
    "Machine translation is amazing!",
    "Transformers have revolutionized NLP.",
    "I love learning new languages."
]

# Step 4: Tokenize the input sentences
inputs = tokenizer(src_texts, return_tensors="pt", padding=True, truncation=True)

# Step 5: Generate translations
translated_tokens = model.generate(**inputs)

# Decode translated tokens to human-readable text
translated_texts = [tokenizer.decode(t, skip_special_tokens=True) for t in translated_tokens]

# Step 6: Print source and translated sentences
for src, tgt in zip(src_texts, translated_texts):
    print(f"Source: {src}")
    print(f"Translation: {tgt}\n")

# Step 7: Evaluate translations using BLEU score
# Load BLEU metric
bleu = evaluate.load("sacrebleu") # Changed from 'bleu = load_metric("sacrebleu")'

# Provide reference translations (list of lists)
references = [
    ["Bonjour, comment ça va ?"],
    ["La traduction automatique est incroyable !"],
    ["Les transformers ont révolutionné le traitement du langage naturel."],
    ["J'adore apprendre de nouvelles langues."]
]

# Compute BLEU score
results = bleu.compute(predictions=translated_texts, references=[[ref] for ref in references])
print("BLEU Score:", results['score'])

# Step 8: Reflection
# Compare translation output with reference sentences.
# Discuss strengths: fluent, context-aware translations
# Discuss challenges: rare words, idiomatic expressions, or complex sentences
```
